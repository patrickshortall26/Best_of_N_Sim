## Experiments

```python
from data_tools import average_experiment
from agentpy_flocking import BoidsModel
import matplotlib.pyplot as plt
import numpy as np
import agentpy as ap
```

## no pooling, no byzantine

```python
# Define parameters
npnb_parameters = {  
    'size': 50,
    'seed': ap.IntRange(0, 10000),
    'steps': 100,
    'ndim': 2,
    'population': 500,
    'inner_radius': 3,
    'outer_radius': 10,
    'border_distance': 10,
    'cohesion_strength': 0.001,
    'seperation_strength': 0.1,
    'alignment_strength': 0.005,
    'border_strength': 0.5,
    'evidence_epsilon' : 0.02,
    'pooling_epsilon' : 0.0,
    'alpha' : 0.1,
    'w' : 1,
    'byzantine_agents' : 0,
    'byzantine_opinion' : 0.01
}

# Get 10 different random seeds 
sample = ap.Sample(npnb_parameters, 10)

# Run experiment and gather results
exp = ap.Experiment(BoidsModel, sample, record=True)
npnb_results = exp.run(-1)
```

```python
# Collect average opinions and average standard deviation accross experiments
mean, std = average_experiment(npnb_results)
t = range(len(mean))

# Create figure
fig, ax = plt.subplots(figsize=(8,8),dpi=800)

# Calculate upper and lower error
min_std = mean - 0.25*std
max_std = mean + 0.25*std

# Plot data
ax.plot(t, mean, 'k')
ax.plot(t, min_std, 'k--', alpha=0.5)
ax.plot(t, max_std, 'k--', alpha=0.5)

# Add Labels
plt.xlabel('Timestep (t)', fontsize=18)
plt.ylabel('Average Opinion (φ)', fontsize=18)

# Adjust axes limits
ax.set_ylim([0.4, 1])

plt.show()
```

## Figures for report
